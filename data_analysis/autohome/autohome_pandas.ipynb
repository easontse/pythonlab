```python
import csv
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

font = FontProperties(fname='/usr/share/fonts/opentype/noto/NotoSansCJK.ttc', size=16)

with open('autohome_count.csv', 'r') as infile:
    reader = csv.reader(infile)
    ftr = ['楼','板凳','楼主','沙发','主楼']
    head30 = []
    for i in reader:
        if len(head30) < 15:
            if i[0] not in ftr:
                head30.append([i[0], int(i[1])])
        else:
            break

data_dict = {x:y for x,y in head30}

df=pd.Series(data_dict).sort_values(ascending=False)
ax = df.plot(kind='bar')

labels = [label for label in df.index.values] 
ax.set_xticklabels(labels, fontproperties=font)
ax.set_title('autohome: 最常出現名詞', fontproperties=font)
```
